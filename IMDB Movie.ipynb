```python
#11 Classify movies based on Ratings(Good,Better,Best)
```

```python
def rating(rating):
    if rating >=8.0:
        return 'Best'
    elif rating >= 6.0:
        return 'Better'
    else:
        return 'Good'
```

```python
data['rates']=data['Rating'].apply(rating)
```
